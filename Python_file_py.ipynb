TASK–4: Conditional Data Handling

```python
data = np.array([12, 15, 20, 20, 25, 30, 35, 40])
print(data)
#Display all values greater than 20
data[data > 20]
```

```python
#Count how many values are less than 20
data[data < 20].size
```

```python
#Replace values less than 20 with 20
data[data < 20] = 20
print(data)
```

TASK–5: Mini Use Case

```python
import numpy as np

data = np.array([12, 15, 20, 20, 25, 30, 35, 40])
print(data)
```

```python
print(np.max(data))
print(np.min(data))
```

```python
#Count Pass Students (pass mark = 20)
passcount = np.sum(data >= 20)
print(passcount)
```

```python
#Average Marks
avg = np.mean(data)
print(avg)
```

```python
#Class Performance
if avg < 20:
    performance = "Poor"
elif avg < 30:
    performance = "Average"
else:
    performance = "Good"
print("Class Performance:", performance)
```
